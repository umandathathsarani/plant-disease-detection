# Notebook 00: Exploratory Data Analysis (EDA)

In this notebook, I will load the PlantDisease dataset and visualize samples to understand the variety of leaf diseases I am dealing with.

### Step 1: Environment Setup and Path Verification
First, I need to set up my environment, check if a GPU is available, and verify that my notebook can correctly locate the extracted dataset.

```python
import os
import torch
import torchvision
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

# Check device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define the path to your dataset (Ensure this matches your folder name)
data_dir = '../data/raw/PlantVillage'

# Check if the path exists
if os.path.exists(data_dir):
    print(f"Dataset path confirmed: {data_dir}")
else:
    print(f"Please ensure the dataset is extracted at: {data_dir}")
```

### Step 2: Visualizing Sample Images

Now that I've confirmed the dataset path, I want to take a look at the actual images. I'll load a few samples from different classes to see what the healthy and diseased leaves look like. This will help me understand the quality of the data I'm working with.

```python
# I'll set up a simple transform to resize the images just for visualization purposes
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor()
])

# Loading the dataset using ImageFolder
dataset = datasets.ImageFolder(data_dir, transform=transform)
classes = dataset.classes

print(f"Total number of classes I'm working with: {len(classes)}")

# Now I'll visualize a few random samples to get a better sense of the data
fig, axes = plt.subplots(1, 4, figsize=(15, 5))
for i in range(4):
    # I'll pick some images from the dataset
    img, label = dataset[i]
    # Converting the tensor to a format matplotlib can display
    axes[i].imshow(img.permute(1, 2, 0))
    axes[i].set_title(classes[label], fontsize=9)
    axes[i].axis('off')

plt.show()
```
